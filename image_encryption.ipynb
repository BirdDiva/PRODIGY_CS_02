```python
from PIL import Image

def encrypt_image(image_path, shift=10):
    try:
        original_image = Image.open(image_path)
        encrypted_image = original_image.point(lambda p: p + shift)
        encrypted_image.save("encrypted_image.png")
        print("Image encrypted successfully!")
    except Exception as e:
        print(f"Error: {e}")

def decrypt_image(encrypted_image_path, shift=10):
    try:
        encrypted_image = Image.open(encrypted_image_path)
        decrypted_image = encrypted_image.point(lambda p: p - shift)
        decrypted_image.save("decrypted_image.png")
        print("Image decrypted successfully!")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    image_path = "/content/encrypt_image.png"
    encrypt_image(image_path)
    decrypt_image("encrypted_image.png")
```
